## 1. Import the Dataset

```python
# Reading the datasets from the CSV files
training_data = pd.read_csv("train.csv")
data_test  = pd.read_csv("test.csv")
```

## 5. Remove the Duplicate Values

```python
# Check the duplicate rows
training_data.duplicated().any()
```

```python
# find the duplicate column and Transpos the DataFrame and check again for duplicate rows
duplicated_columns = training_data.columns[training_data.T.duplicated()].tolist()
```

```python
# length of the duplicat column
len(duplicated_columns)
```

```python
# Drop the duplicate columns
training_data = training_data.drop(duplicated_columns,axis=1)
```

```python
# Check missing values 
training_data.isnull().sum()
```

```python
# import the libraries for the visualization or graph
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plot the count plot to visualize the distribution of values
sns.countplot(x=training_data['Activity'])
# Rotate the x-axis labels by 35 degrees
plt.xticks(rotation=35)
# Display the plot
plt.show()
```

## 7. Store the Feature Matrix

```python
# Split the dataset into features 'x' and target variable 'y', with the 'Activity'
X = training_data.drop('Activity',axis=1)
y = training_data['Activity']
```

```python
y
```

```python
# Import the LabelEncoder from the sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder
# Create the variable of the  LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

## 8. Split The Dataset

```python
# Import the train_test_split function from the sklearn.model_selection to split the data
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into training and testing sets 80% for training and 20% for testing
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20, random_state=42)
```

## 9. Logistic Regression

```python
# import the libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Create variable for the LogisticRegression model with 500 iterations
log = LogisticRegression(max_iter=500)
# Train the model 
log.fit(X_train, y_train)
```

```python
# Predict the target values for the test dataset
y_pred1 = log.predict(X_test)
# Calculate the accuracy score
accuracy_score(y_test,y_pred1)
```

## 10. Random Forest Classifier

```python
# Import the Random Forest Classifier from sklearn.ensemble
from sklearn.ensemble import RandomForestClassifier
# Create the variable for RandomForestClassifier
RF = RandomForestClassifier()
# Train the random forest model on the dataset
RF.fit(X_train,y_train)
```

```python
# again Predict the target values for the test dataset
y_pred2 = RF.predict(X_test)
# Calculate the accuracy score
accuracy_score(y_test,y_pred2)
```

## 11. Feature Selection

###  Filter Method

```python
# Import SelectKBest and f_classif from sklearn.feature_selection
from sklearn.feature_selection import SelectKBest,f_classif
```

```python
# Set the number k=200
k=200

# Initialize feature selector
selector = SelectKBest(f_classif,k=k)

# Feature selector to fit on the training data
X_train_selected = selector.fit_transform(X_train,y_train)

# Transform the test dataset
X_test_selected = selector.transform(X_test)

# Get the indices from the selected features
selected_indices=selector.get_support(indices=True)
selected_features = X_train.columns[selected_indices]

# Print the length of selected features
print(len(selected_features))
```

### Wrapper Method

```python
# Import Recursive Feature Elimination (RFE) from sklearn.feature_selection
from sklearn.feature_selection import RFE
# Import RandomForestClassifier from sklearn.ensemble
from sklearn.ensemble import RandomForestClassifier
```

```python
estimator = RandomForestClassifier()
```

```python
# Set the number k=100 for RFE
k=100
rfe_selector = RFE(estimator,n_features_to_select=k)

# Transform the train and test dataset to keep the selected features after RFE
X_train_selected_rfe = rfe_selector.fit_transform(X_train_selected,y_train)
X_test_selected_rfe = rfe_selector.transform(X_test_selected)

# Get the indices from the selected features
selected_indices_rfe = rfe_selector.get_support(indices=True)

# Get the name from selected features by indices
selected_features_rfe = selected_features[selected_indices_rfe]
print(selected_features_rfe)
```

```python
# Print the length 
print(len(selected_features_rfe))
```

```python
RF = RandomForestClassifier()
```

```python
# Fit the RF model
RF.fit(X_train_selected_rfe,y_train)
```

```python
# Predict the target values
y_pred_rf = RF.predict(X_test_selected_rfe)
```

```python
# Import the accuracy_score from sklearn.metrics
from sklearn.metrics import accuracy_score
```

```python
# Calculate the accuracy score
accuracy_score(y_test,y_pred_rf)
```
